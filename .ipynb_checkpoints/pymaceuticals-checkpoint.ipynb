```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "Instructions/Pymaceuticals/data/mouse_drug_data.csv"
clinical_trial_data_to_load = "Instructions/Pymaceuticals/data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
ct_df = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
full_df = pd.merge(mouse_df, ct_df, how='left', on='Mouse ID')

# Display the data table for preview
full_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
jac = full_df.drop(['Metastatic Sites', 'Mouse ID'], axis=1)

mean_vol_0 = jac.groupby(['Drug','Timepoint'])
mean_vol = mean_vol_0.mean()
mean_vol
```

```python
# Convert to DataFrame
mv_df = mean_vol.reset_index()

# Preview DataFrame
mv_df.head()

mv_dff = mv_df.pivot(index='Timepoint', columns='Drug')

mv_dff.columns = mv_dff.columns.droplevel(0)
mv_dff

mv_col = mv_dff.columns

mv_dff
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
jac1 = full_df.drop(['Metastatic Sites', 'Mouse ID'], axis=1)

std_vol_0 = jac1.groupby(['Drug','Timepoint'])
std_vol = std_vol_0.sem()

# Convert to DataFrame
sev_df = std_vol.reset_index()

# Preview DataFrame
sev_df
```

```python
sev_dff = sev_df.pivot(index='Timepoint', columns='Drug')
sev_dff.columns = sev_dff.columns.droplevel(0)
sev_dff
```

```python
columns = sev_dff.columns
```

```python
# ax.errorbar(sev_dff.index, mv_dff['Capomulin'], yerr=sev_dff['Capomulin'], fmt="o", color="b",
#             alpha=0.5, label="Capomulin")
# ax.errorbar(sev_dff.index, mv_dff['Ceftamin'], yerr=sev_dff['Ceftamin'], fmt="o", color="r",
#             alpha=0.5, label="Ceftamin")

# ax.set_xlim(-0.5, len(means))
```

```python
# Generate the Plot (with Error Bars)
# Plot sample means with error bars
def volume_plot():
    fig, ax = plt.subplots()
    for x in columns:
        ax.errorbar(sev_dff.index, mv_dff[x], yerr=sev_dff[x], fmt="o", alpha=0.5, label=x)

    ax.set_xlabel("Timepoint")
    ax.set_ylabel("Mean Volume of Tumor (mm3)")

    plt.legend(loc=2, fontsize="small", prop={'size': 7}, fancybox=False)

    plt.savefig("Tumor Response Scatter")
    plt.show()

volume_plot()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
jac = full_df.drop(['Tumor Volume (mm3)', 'Mouse ID'], axis=1)

mtst_gp = jac.groupby(['Drug','Timepoint'])
mtst_mean = mtst_gp.mean()
mtst_mean

# Convert to DataFrame
mtst_df = mtst_mean.reset_index()

# Preview DataFrame
mtst_df = mtst_df.pivot(index='Timepoint', columns='Drug')
mtst_df.columns = mtst_df.columns.droplevel(0)
mtst_df
```

```python
# Generate the Plot (Accounting for percentages)

def survival_plot():
    fig, ax = plt.subplots()
    for x in columns:
        ax.plot(mtst_df.index, mtst_df[x])
#     ax.scatter(moco_dff.index, moco_dff['Capomulin'])
#     ax.scatter(moco_dff.index, moco_dff['Ceftamin'])

    ax.set_xlim(-0.5, 48)
    ax.set_xlabel("Timepoint")
    ax.set_ylabel("Survival Rate")

    plt.legend(labels=columns, loc="best", fontsize="small", fancybox=True)

    plt.savefig("Survival Scatter")
    plt.show()
    
survival_plot()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mid = full_df[["Mouse ID", "Timepoint", "Drug"]]
mouse_count = mid.groupby(['Drug', 'Timepoint']).count()
# Convert to DataFrame

# Preview DataFrame
mouse_count.head()
```

```python
promoco_df = mouse_count.reset_index()
promoco_df
```

```python
# Minor Data Munging to Re-Format the Data Frames
moco_df = promoco_df.pivot(index='Timepoint', columns='Drug')
moco_df.columns = moco_df.columns.droplevel(0)

# Preview the Data Frame
moco_df
```

```python
columns = ['Capomulin', 'Ceftamin', 'Infubinol', 'Ketapril', 'Naftisol', 'Placebo',
       'Propriva', 'Ramicane', 'Stelasyn', 'Zoniferol']
columns
```

```python
# # create function that performs the percent calculation across all coulumns
def survivalrate():
    for x in columns:
        moco_df[x] = moco_df[x]/moco_df[x][0]
survivalrate()
perct_moco = moco_df
```

```python
perct_moco = perct_moco.round(2)
perct_moco
```

```python
# Generate the Plot (Accounting for percentages)

def survival_plot():
    fig, ax = plt.subplots()
    for x in columns:
        ax.plot(perct_moco.index, perct_moco[x])
#     ax.scatter(moco_dff.index, moco_dff['Capomulin'])
#     ax.scatter(moco_dff.index, moco_dff['Ceftamin'])

    ax.set_xlim(-0.5, 48)
    ax.set_xlabel("Timepoint")
    ax.set_ylabel("Survival Rate")

    plt.legend(labels=columns, loc="best", fontsize="small", fancybox=True)

    plt.savefig("Survival Scatter")
    plt.show()
    
survival_plot()

```
